```python
import pandas as pd
from decimal import Decimal
```

```python
pd.options.display.max_colwidth=None
pd.options.display.max_rows =None
```

```python
df = pd.read_csv("pp.csv").iloc[:,1:]
```

```python
msgs = []
for k in range(len(df)):
    df.iloc[k,2] = round(df.iloc[k,2],2) 
    df.iloc[k,3] = Decimal(str(df.iloc[k,3])).quantize(Decimal(".01")) 
    df.iloc[k,5] = round(float("0."+str(int(df.iloc[k,5]))),4)
    msg = ""
    for h in df.iloc[k,4]:
        msg+=harbour_law[h]
        if df.iloc[k,4].index(h) != len(df.iloc[k,4])-1:
            msg+=" "
    msgs.append(msg)
del df["comment"]
df["message"]=msgs
```
